# Univariate analysis

```python
# One molecule
from rdkit import Chem
from rdkit.Chem import Descriptors 

sample_molecule = Chem.MolFromSmiles(ligands_subset["Ligand SMILES"].iloc[0])

print("Molecular weight of the first molecule: ", Descriptors.MolWt(sample_molecule))
print("Number of valence electrons of the first molecule: ", Descriptors.NumValenceElectrons(sample_molecule))
Chem.Descriptors.CalcMolDescriptors(sample_molecule, missingVal=None, silent=True)
```

```python
# 1 Molecular weight
import time
import numpy as np

ligangs_as_molecules = []

# Liste pour stocker les temps d'exécution
execution_times = []
molecular_weights = []

for ligand in ligands_subset["Ligand SMILES"]:
    start_time = time.time()
    molecule = Chem.MolFromSmiles(ligand)
    mw = Descriptors.MolWt(molecule)   
    molecular_weights.append(mw) 
    end_time = time.time()
    execution_times.append(end_time - start_time)

    # Calculer les statistiques
mean_time = np.mean(execution_times)
std_time = np.std(execution_times)

print(f"Mean execution time: {mean_time} seconds")
print(f"Standard deviation: {std_time} seconds")
print(f"Estimation for the full dataset: {round(len(df)*mean_time/60)} minutes")
```

```python
import pandas as pd
molecular_weights = pd.Series(molecular_weights)
molecular_weights.plot(kind = 'hist')
```

```python
# Check if any atoms are aromatic
is_aromatic = any(atom.GetIsAromatic() for atom in sample_molecule.GetAtoms())

print("Aromatic" if is_aromatic else "Aliphatic")
```

```python
# Liste pour stocker les temps d'exécution
execution_times = []
are_aromatic = []

for ligand in ligands_subset["Ligand SMILES"]:
    start_time = time.time()
    molecule = Chem.MolFromSmiles(ligand)
    try:
        is_aromatic = any(atom.GetIsAromatic() for atom in molecule.GetAtoms())
    except:
        pass
    are_aromatic.append(is_aromatic)
    end_time = time.time()
    execution_times.append(end_time - start_time)

    # Calculer les statistiques
mean_time = np.mean(execution_times)
std_time = np.std(execution_times)

print(f"Mean execution time: {mean_time} seconds")
print(f"Standard deviation: {std_time} seconds")
print(f"Estimation for the full dataset: {round(len(df)*mean_time/60)} minutes")

print(f"{np.mean(are_aromatic)} are aromatic")
```

```python
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors

# Définir les motifs des groupes fonctionnels
functional_groups = {
    "Hydroxyl": Chem.MolFromSmarts("[OX2H]"),
    "Carbonyl": Chem.MolFromSmarts("[CX3]=[OX1]"),
    "Amine": Chem.MolFromSmarts("[NX3;H2,H1;!$(NC=O)]"),
    "Carboxyl": Chem.MolFromSmarts("C(=O)[OX2H1]"),
    "Benzene": Chem.MolFromSmarts("c1ccccc1")
}

# Exemple de liste de SMILES
ligands = ligands_subset['Ligand SMILES']

# Parcourir les ligands et identifier les groupes fonctionnels
for ligand in ligands:
    molecule = Chem.MolFromSmiles(ligand)
    print(f"Ligand: {ligand}")
    for group_name, group_smarts in functional_groups.items():
        if molecule.HasSubstructMatch(group_smarts):
            print(f"  Contains {group_name}")
    print()
```
